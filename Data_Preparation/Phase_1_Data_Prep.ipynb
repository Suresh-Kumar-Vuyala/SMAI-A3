### Step 4: Filter & Split (Street Food Selection)

```python
output_dir = os.path.join(base_path, 'dataset_split')

# Structure check
search_path = os.path.join(raw_extract_path, 'Indian Food Images', 'Indian Food Images')
if not os.path.exists(search_path): search_path = raw_extract_path

street_food = [
    'aloo_tikki', 'bhatura', 'chana_masala', 'daal_puri',
    'jalebi', 'kachori', 'lassi', 'litti_chokha',
    'poha', 'rabri', 'kuzhi_paniyaram', 'unni_appam'
]

for cls in street_food:
    src = os.path.join(search_path, cls)
    if not os.path.exists(src): continue

    imgs = [f for f in os.listdir(src) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    random.shuffle(imgs)

    train_idx, val_idx = int(len(imgs)*0.7), int(len(imgs)*0.85)

    for name, split in [('train', imgs[:train_idx]), ('val', imgs[train_idx:val_idx]), ('test', imgs[val_idx:])]:
        dst = os.path.join(output_dir, name, cls.replace(' ', '_'))
        os.makedirs(dst, exist_ok=True)
        for img in split:
            shutil.copy(os.path.join(src, img), os.path.join(dst, img))
    print(f"✅ {cls} processed.")

print("\n🚀 READY! Dataset located at:", output_dir)
```

```python
print("{:<20} | {:<7} | {:<7} | {:<7} | {:<7}".format("Food Class", "Train", "Val", "Test", "Total"))
print("-" * 55)

grand_total = 0
for cls in street_food:
    counts = []
    for split in ['train', 'val', 'test']:
        path = os.path.join(output_dir, split, cls)
        count = len(os.listdir(path)) if os.path.exists(path) else 0
        counts.append(count)

    total = sum(counts)
    grand_total += total
    print("{:<20} | {:<7} | {:<7} | {:<7} | {:<7}".format(cls, counts[0], counts[1], counts[2], total))

print("-" * 55)
print(f"GRAND TOTAL IMAGES: {grand_total}")
```
