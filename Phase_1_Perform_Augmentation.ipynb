# Phase 1: Physical Data Augmentation (Saving to Drive)
This notebook will read your existing training images and create **2 additional augmented versions** for each one, effectively tripling your training data.

```python
from google.colab import drive
import os, random
from PIL import Image
from torchvision import transforms

drive.mount('/content/drive')
```

```python
# Configuration
base_path = '/content/drive/MyDrive/SMAI_Assignment3/dataset_split/train'
n_aug_per_image = 2 # How many new versions to create per image

# Define what the augmentation looks like
aug_pipeline = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(degrees=15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2)
])

classes = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

def perform_offline_augmentation():
    print(f"🚀 Starting augmentation for {len(classes)} classes...")

    for cls in classes:
        cls_path = os.path.join(base_path, cls)
        # Only augment the original images (avoid re-augmenting already augmented ones)
        original_images = [f for f in os.listdir(cls_path) if not "_aug" in f]

        print(f"Processing {cls}: {len(original_images)} images found.")

        for img_name in original_images:
            img_path = os.path.join(cls_path, img_name)
            try:
                img = Image.open(img_path).convert('RGB')
                basename, ext = os.path.splitext(img_name)

                for i in range(1, n_aug_per_image + 1):
                    aug_img = aug_pipeline(img)
                    save_name = f"{basename}_aug{i}{ext}"
                    aug_img.save(os.path.join(cls_path, save_name))
            except Exception as e:
                print(f"❌ Error processing {img_name}: {e}")

        total_count = len(os.listdir(cls_path))
        print(f"✅ {cls} finished. New total: {total_count} images.")

perform_offline_augmentation()
print("\n✨ Dataset tripling complete!")
```

```python
# Redefining paths to avoid NameError
parent_dir = '/content/drive/MyDrive/SMAI_Assignment3/dataset_split'
train_dir = os.path.join(parent_dir, 'train')
classes = sorted([d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))]) if os.path.exists(train_dir) else []

header = "{:<20} | {:<12} | {:<12} | {:<7} | {:<7} | {:<7}".format("Food Class", "Train(Orig)", "Train(Aug)", "Val", "Test", "Total")
print(header)
print("-" * len(header))

grand_total = 0
for cls in classes:
    # Train counts
    tr_path = os.path.join(train_dir, cls)
    all_train = os.listdir(tr_path) if os.path.exists(tr_path) else []
    tr_orig = len([f for f in all_train if "_aug" not in f])
    tr_aug = len([f for f in all_train if "_aug" in f])

    # Val count
    val_path = os.path.join(parent_dir, 'val', cls)
    val_count = len(os.listdir(val_path)) if os.path.exists(val_path) else 0

    # Test count
    test_path = os.path.join(parent_dir, 'test', cls)
    test_count = len(os.listdir(test_path)) if os.path.exists(test_path) else 0

    total = tr_orig + tr_aug + val_count + test_count
    grand_total += total

    print("{:<20} | {:<12} | {:<12} | {:<7} | {:<7} | {:<7}".format(cls, tr_orig, tr_aug, val_count, test_count, total))

print("-" * len(header))
print(f"TOTAL SYSTEM IMAGES: {grand_total}")
```
